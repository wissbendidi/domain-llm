```python
# Domain LLM v1.1 - Improved Model Training
# AI Engineer Homework - Systematic Model Improvement
# This notebook addresses the critical issues identified in baseline v1.0

import os
import json
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer,
    DataCollatorForLanguageModeling,
    EarlyStoppingCallback
)
from peft import LoraConfig, get_peft_model, TaskType
from google.colab import drive, files
import re
from typing import List, Dict, Tuple
import time
from datetime import datetime

# ================================================================
# 1. ENVIRONMENT SETUP & CONFIGURATION
# ================================================================

print("🚀 Domain LLM v1.1 - Improved Model Training")
print("=" * 60)
print("📊 Addressing baseline issues identified:")
print("   - Generation artifacts (<|endoftext|> tokens)")
print("   - Low domain validity (4% → Target: 80%+)")
print("   - No safety filtering")
print("   - Poor semantic relevance")
print("=" * 60)

# Install required packages
!pip install transformers datasets accelerate peft bitsandbytes wandb scikit-learn

# Mount Google Drive
print("📂 Mounting Google Drive...")
drive.mount('/content/drive')

# Set up device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🖥️ Using device: {device}")
print(f"🔧 GPU: {torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'Not available'}")

# ================================================================
# 2. IMPROVED CONFIGURATION - ADDRESSING BASELINE LIMITATIONS
# ================================================================

class ModelConfig:
    """Improved configuration addressing baseline limitations"""

    # Model Configuration
    BASE_MODEL = "TinyLlama/TinyLlama-1.1B-intermediate-step-1431k-3T"
    MODEL_VERSION = "v1.1"
    OUTPUT_DIR = f"/content/tinyllama-domain-generator-{MODEL_VERSION}"

    # Training Parameters - IMPROVED from baseline
    NUM_EPOCHS = 3              # Increased from 1
    BATCH_SIZE = 4              # Increased from 1
    GRADIENT_ACCUMULATION = 4   # Increased from 1
    LEARNING_RATE = 2e-4        # Optimized from 5e-5
    WARMUP_STEPS = 50           # Added warmup (was 0)
    MAX_LENGTH = 128            # Same as baseline
    WEIGHT_DECAY = 0.01         # Added regularization

    # LoRA Configuration - ENHANCED
    LORA_R = 16                 # Increased from 8
    LORA_ALPHA = 32             # Increased from 16
    LORA_DROPOUT = 0.1          # Added dropout (was 0.0)
    LORA_TARGET_MODULES = [     # Expanded from just q_proj, v_proj
        "q_proj", "v_proj", "k_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj"
    ]

    # Data Configuration
    VALIDATION_SPLIT = 0.2      # Added validation (baseline had none)

    # Generation Parameters - IMPROVED
    GENERATION_CONFIG = {
        "temperature": 0.7,      # Added controlled randomness
        "do_sample": True,       # Enabled sampling (was False)
        "top_p": 0.9,           # Added nucleus sampling
        "top_k": 50,            # Added top-k filtering
        "repetition_penalty": 1.1,  # Reduce repetition
        "early_stopping": True   # Better stopping
    }

config = ModelConfig()

print(f"📋 MODEL CONFIGURATION v{config.MODEL_VERSION}")
print(f"   Base Model: {config.BASE_MODEL}")
print(f"   Training Epochs: {config.NUM_EPOCHS} (vs 1 in baseline)")
print(f"   Batch Size: {config.BATCH_SIZE} (vs 1 in baseline)")
print(f"   Learning Rate: {config.LEARNING_RATE} (vs 5e-5 in baseline)")
print(f"   LoRA Rank: {config.LORA_R} (vs 8 in baseline)")
print(f"   Validation Split: {config.VALIDATION_SPLIT} (vs None in baseline)")

# ================================================================
# 3. ENHANCED DATA LOADING & PREPROCESSING
# ================================================================

def upload_training_data():
    """Upload and validate training data"""
    print("\n📂 Upload your improved training dataset:")
    print("   - Should be JSONL format")
    print("   - Each line: {'prompt': 'business description', 'completion': 'domain.com'}")
    print("   - Recommended: 500+ examples for better performance")

    uploaded = files.upload()

    if uploaded:
        filename = list(uploaded.keys())[0]
        file_path = f"/content/{filename}"

        # Validate file
        file_size = os.path.getsize(file_path)
        print(f"✅ File uploaded: {filename}")
        print(f"📊 File size: {file_size:,} bytes")

        return file_path
    else:
        print("⚠️ No file uploaded. Using fallback data...")
        return None

def validate_and_clean_dataset(file_path: str) -> List[Dict]:
    """Enhanced data validation and cleaning"""

    if file_path is None:
        # Fallback: Create enhanced sample dataset
        print("🔧 Creating enhanced sample dataset...")
        return create_enhanced_sample_data()

    print(f"🔍 Validating dataset: {file_path}")

    data = []
    invalid_count = 0

    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            for line_num, line in enumerate(f, 1):
                try:
                    item = json.loads(line.strip())

                    # Enhanced validation
                    if not isinstance(item, dict):
                        invalid_count += 1
                        continue

                    if 'prompt' not in item or 'completion' not in item:
                        invalid_count += 1
                        continue

                    # Clean and validate prompt
                    prompt = str(item['prompt']).strip()
                    completion = str(item['completion']).strip()

                    if len(prompt) < 5 or len(completion) < 5:
                        invalid_count += 1
                        continue

                    # Validate domain format
                    if not is_valid_domain_format(completion):
                        invalid_count += 1
                        continue

                    # Apply content safety filter
                    if not is_safe_content(prompt):
                        invalid_count += 1
                        print(f"⚠️ Filtered unsafe content on line {line_num}")
                        continue

                    data.append({
                        'prompt': prompt,
                        'completion': completion
                    })

                except json.JSONDecodeError:
                    invalid_count += 1
                    continue

        print(f"✅ Dataset validation complete:")
        print(f"   Valid examples: {len(data)}")
        print(f"   Invalid/filtered: {invalid_count}")
        print(f"   Quality rate: {len(data)/(len(data)+invalid_count)*100:.1f}%")

        return data

    except Exception as e:
        print(f"❌ Error loading dataset: {e}")
        return create_enhanced_sample_data()

def is_valid_domain_format(domain: str) -> bool:
    """Validate domain format"""
    valid_extensions = ['.com', '.net', '.org', '.io', '.co', '.ai', '.app']
    return (
        any(domain.endswith(ext) for ext in valid_extensions) and
        len(domain) >= 5 and len(domain) <= 50 and
        ' ' not in domain and
        domain.count('.') == 1
    )

def is_safe_content(prompt: str) -> bool:
    """Basic content safety filter"""
    unsafe_keywords = [
        'adult', 'explicit', 'porn', 'nude', 'sex',
        'gambling', 'casino', 'betting',
        'weapons', 'guns', 'firearms',
        'drugs', 'cocaine', 'marijuana',
        'hate', 'racist', 'nazi',
        'violence', 'kill', 'murder'
    ]

    prompt_lower = prompt.lower()
    return not any(keyword in prompt_lower for keyword in unsafe_keywords)

def create_enhanced_sample_data() -> List[Dict]:
    """Create enhanced sample dataset with better quality"""

    enhanced_data = [
        # Tech/AI businesses
        {"prompt": "an AI assistant for coding", "completion": "codecompanion.ai"},
        {"prompt": "a machine learning platform for healthcare", "completion": "medailytics.com"},
        {"prompt": "a cloud-based data analytics tool", "completion": "cloudmetrics.io"},
        {"prompt": "an automation platform for small businesses", "completion": "automate.co"},
        {"prompt": "a cybersecurity monitoring service", "completion": "shieldwatch.com"},

        # E-commerce/Retail
        {"prompt": "an online marketplace for vintage items", "completion": "vintagevault.com"},
        {"prompt": "a subscription box for pet supplies", "completion": "pawsbox.io"},
        {"prompt": "a sustainable fashion marketplace", "completion": "ecofashion.co"},
        {"prompt": "a personalized nutrition planning app", "completion": "nutriplan.app"},

        # Local Services
        {"prompt": "a yoga studio specializing in hot yoga", "completion": "hotyogi.com"},
        {"prompt": "an organic coffee roastery", "completion": "purebeans.co"},
        {"prompt": "a mobile car detailing service", "completion": "sparklewheels.com"},
        {"prompt": "a boutique bakery with custom cakes", "completion": "sweetcreations.co"},

        # Creative/Media
        {"prompt": "a podcast production company", "completion": "audiocrafters.com"},
        {"prompt": "a freelance graphic design studio", "completion": "pixelstudio.co"},
        {"prompt": "a wedding photography business", "completion": "momentcapture.com"},

        # Professional Services
        {"prompt": "a digital marketing consultancy", "completion": "growthexperts.io"},
        {"prompt": "a financial planning advisory", "completion": "wealthpath.com"},
        {"prompt": "a legal services firm for startups", "completion": "startuplaw.co"},

        # Health/Wellness
        {"prompt": "a mental health counseling practice", "completion": "mindwellness.com"},
        {"prompt": "a fitness coaching platform", "completion": "fitcoach.app"},
        {"prompt": "a meditation and mindfulness app", "completion": "peacemind.io"}
    ]

    print(f"📝 Using enhanced sample dataset with {len(enhanced_data)} examples")
    return enhanced_data

def create_improved_training_prompt(prompt: str, completion: str) -> str:
    """Improved prompt formatting for better training"""
    return f"Business: {prompt}\nDomain: {completion}<|endoftext|>"

def create_train_validation_split(data: List[Dict], validation_ratio: float = 0.2) -> Tuple[List[Dict], List[Dict]]:
    """Create train/validation split with stratification"""

    # Shuffle data
    import random
    random.seed(42)  # For reproducibility
    shuffled_data = data.copy()
    random.shuffle(shuffled_data)

    # Split
    split_idx = int(len(shuffled_data) * (1 - validation_ratio))
    train_data = shuffled_data[:split_idx]
    val_data = shuffled_data[split_idx:]

    print(f"📊 Data split created:")
    print(f"   Training examples: {len(train_data)}")
    print(f"   Validation examples: {len(val_data)}")
    print(f"   Split ratio: {len(train_data)/(len(train_data)+len(val_data))*100:.1f}% train")

    return train_data, val_data

# ================================================================
# 4. LOAD AND PREPARE DATA
# ================================================================

print("\n" + "="*60)
print("📊 DATA LOADING & PREPARATION")
print("="*60)

# Upload data
training_file = upload_training_data()

# Load and validate
raw_data = validate_and_clean_dataset(training_file)

if len(raw_data) < 10:
    print("⚠️ Warning: Very small dataset. Consider adding more examples for better performance.")

# Create train/validation split
train_data, val_data = create_train_validation_split(raw_data, config.VALIDATION_SPLIT)

# Preprocess data
def preprocess_dataset(data: List[Dict]) -> List[Dict]:
    """Preprocess data for training"""
    processed = []
    for item in data:
        training_text = create_improved_training_prompt(item['prompt'], item['completion'])
        processed.append({'text': training_text})
    return processed

train_processed = preprocess_dataset(train_data)
val_processed = preprocess_dataset(val_data)

# Create datasets
train_dataset = Dataset.from_list(train_processed)
val_dataset = Dataset.from_list(val_processed)

# Display sample training examples
print("\n👀 Sample training examples:")
for i in range(min(3, len(train_processed))):
    print(f"\nExample {i+1}:")
    print(f"Original: {train_data[i]['prompt']} → {train_data[i]['completion']}")
    print(f"Training format: {train_processed[i]['text'][:100]}...")

# ================================================================
# 5. MODEL LOADING & CONFIGURATION
# ================================================================

print("\n" + "="*60)
print("🤖 MODEL LOADING & CONFIGURATION")
print("="*60)

# Load tokenizer
print("📥 Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(config.BASE_MODEL)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

# Load base model
print("📥 Loading base model...")
model = AutoModelForCausalLM.from_pretrained(
    config.BASE_MODEL,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True
)

# Configure improved LoRA
print("🔧 Configuring enhanced LoRA...")
lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=config.LORA_R,
    lora_alpha=config.LORA_ALPHA,
    lora_dropout=config.LORA_DROPOUT,
    target_modules=config.LORA_TARGET_MODULES
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

print(f"✅ Model configuration complete:")
print(f"   Trainable parameters: {model.num_parameters():,}")
print(f"   LoRA rank: {config.LORA_R}")
print(f"   Target modules: {len(config.LORA_TARGET_MODULES)}")

# ================================================================
# 6. TOKENIZATION - FIXED for proper batching
# ================================================================

def tokenize_function(examples):
    """Enhanced tokenization function with proper padding and truncation"""
    texts = examples["text"] if isinstance(examples["text"], list) else [examples["text"]]

    # Tokenize with padding and truncation to ensure consistent lengths
    tokenized = tokenizer(
        texts,
        truncation=True,
        padding="max_length",  # Changed from False to "max_length"
        max_length=config.MAX_LENGTH,
        return_tensors=None
    )

    # Copy input_ids to labels for language modeling
    tokenized["labels"] = tokenized["input_ids"].copy()
    return tokenized

print("\n🔤 Tokenizing datasets...")
tokenized_train = train_dataset.map(
    tokenize_function,
    batched=True,
    remove_columns=train_dataset.column_names
)

tokenized_val = val_dataset.map(
    tokenize_function,
    batched=True,
    remove_columns=val_dataset.column_names
)

print(f"✅ Tokenization complete:")
print(f"   Train dataset: {len(tokenized_train)} examples")
print(f"   Validation dataset: {len(tokenized_val)} examples")

# Verify tokenization worked properly
sample_item = tokenized_train[0]
print(f"   Sample input length: {len(sample_item['input_ids'])}")
print(f"   Sample labels length: {len(sample_item['labels'])}")

# ================================================================
# 7. TRAINING CONFIGURATION - FIXED data collator
# ================================================================

print("\n" + "="*60)
print("🎯 TRAINING CONFIGURATION")
print("="*60)

# Fixed data collator - simpler configuration
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False,  # Causal language modeling
    pad_to_multiple_of=None  # Removed pad_to_multiple_of which can cause issues
)

# Enhanced training arguments - FIXED for compatibility
training_args = TrainingArguments(
    output_dir=config.OUTPUT_DIR,

    # Training parameters
    num_train_epochs=config.NUM_EPOCHS,
    per_device_train_batch_size=config.BATCH_SIZE,
    per_device_eval_batch_size=config.BATCH_SIZE,
    gradient_accumulation_steps=config.GRADIENT_ACCUMULATION,

    # Optimization
    learning_rate=config.LEARNING_RATE,
    warmup_steps=config.WARMUP_STEPS,
    weight_decay=config.WEIGHT_DECAY,

    # Evaluation & Saving - FIXED parameter names
    eval_strategy="steps",        # Changed from 'evaluation_strategy'
    eval_steps=50,
    save_strategy="steps",
    save_steps=50,
    save_total_limit=3,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,

    # Logging
    logging_steps=25,
    logging_dir=f"{config.OUTPUT_DIR}/logs",

    # Performance
    fp16=True,
    dataloader_pin_memory=False,
    remove_unused_columns=False,

    # Reporting
    report_to="none",
    run_name=f"domain-llm-{config.MODEL_VERSION}"
)

# Initialize trainer with enhanced configuration
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    data_collator=data_collator,
    tokenizer=tokenizer,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=3)]
)

print(f"📋 Training configuration:")
print(f"   Epochs: {config.NUM_EPOCHS}")
print(f"   Effective batch size: {config.BATCH_SIZE * config.GRADIENT_ACCUMULATION}")
print(f"   Learning rate: {config.LEARNING_RATE}")
print(f"   Warmup steps: {config.WARMUP_STEPS}")
print(f"   Early stopping: Enabled (patience=3)")

# ================================================================
# 8. TRAINING EXECUTION
# ================================================================

print("\n" + "="*60)
print("🚀 MODEL TRAINING - IMPROVED VERSION")
print("="*60)

print(f"🎯 Training objectives:")
print(f"   1. Fix generation artifacts (target: 0% <|endoftext|> tokens)")
print(f"   2. Improve domain validity (target: 80%+ vs 4% baseline)")
print(f"   3. Enhance semantic relevance")
print(f"   4. Add proper validation monitoring")

print(f"\n⏰ Estimated training time: ~{config.NUM_EPOCHS * 2}-{config.NUM_EPOCHS * 3} minutes")
print(f"💾 Model will be saved to: {config.OUTPUT_DIR}")

start_time = time.time()

try:
    # Start training
    print("\n🏃‍♂️ Starting training...")
    trainer.train()

    training_time = time.time() - start_time
    print(f"\n✅ Training completed successfully!")
    print(f"⏰ Total training time: {training_time/60:.1f} minutes")

except Exception as e:
    print(f"\n❌ Training failed: {e}")
    raise

# ================================================================
# 9. MODEL SAVING & VERSIONING
# ================================================================

print("\n" + "="*60)
print("💾 MODEL SAVING & VERSIONING")
print("="*60)

# Save the model
print("💾 Saving trained model...")
trainer.save_model()
tokenizer.save_pretrained(config.OUTPUT_DIR)

# Create model metadata
model_metadata = {
    "model_version": config.MODEL_VERSION,
    "base_model": config.BASE_MODEL,
    "training_date": datetime.now().isoformat(),
    "training_time_minutes": (time.time() - start_time) / 60,

    # Training configuration
    "training_config": {
        "num_epochs": config.NUM_EPOCHS,
        "batch_size": config.BATCH_SIZE,
        "learning_rate": config.LEARNING_RATE,
        "lora_r": config.LORA_R,
        "lora_alpha": config.LORA_ALPHA,
        "validation_split": config.VALIDATION_SPLIT
    },

    # Data information
    "data_info": {
        "train_examples": len(train_data),
        "val_examples": len(val_data),
        "total_examples": len(raw_data)
    },

    # Training results
    "training_results": {
        "final_train_loss": trainer.state.log_history[-2].get('train_loss') if len(trainer.state.log_history) > 1 else None,
        "final_eval_loss": trainer.state.log_history[-1].get('eval_loss') if trainer.state.log_history else None,
        "best_eval_loss": min([log.get('eval_loss', float('inf')) for log in trainer.state.log_history if 'eval_loss' in log], default=None)
    },

    # Improvements over baseline
    "improvements_over_baseline": [
        "Multi-epoch training (3 vs 1)",
        "Validation set monitoring",
        "Enhanced LoRA configuration",
        "Improved generation parameters",
        "Better data preprocessing",
        "Early stopping implementation"
    ]
}

# Save metadata
metadata_path = f"{config.OUTPUT_DIR}/model_metadata_{config.MODEL_VERSION}.json"
with open(metadata_path, 'w') as f:
    json.dump(model_metadata, f, indent=2)

print(f"✅ Model saved successfully:")
print(f"   📂 Model: {config.OUTPUT_DIR}")
print(f"   📄 Metadata: {metadata_path}")

# ================================================================
# 10. IMPROVED DOMAIN GENERATION FUNCTION
# ================================================================

class ImprovedDomainGenerator:
    """Enhanced domain generator addressing baseline issues"""

    def __init__(self, model, tokenizer):
        self.model = model
        self.tokenizer = tokenizer
        self.generation_config = config.GENERATION_CONFIG

    def generate_domain(self, business_description: str, max_attempts: int = 3) -> str:
        """Generate domain with improved parameters and post-processing"""

        # Apply safety filter first
        if not is_safe_content(business_description):
            return "BLOCKED: Inappropriate content"

        prompt = f"Business: {business_description}\nDomain:"

        for attempt in range(max_attempts):
            try:
                inputs = self.tokenizer(prompt, return_tensors="pt", padding=True)
                device = next(self.model.parameters()).device
                inputs = {k: v.to(device) for k, v in inputs.items()}

                with torch.no_grad():
                    outputs = self.model.generate(
                        **inputs,
                        max_length=len(inputs['input_ids'][0]) + 20,
                        temperature=self.generation_config['temperature'],
                        do_sample=self.generation_config['do_sample'],
                        top_p=self.generation_config['top_p'],
                        top_k=self.generation_config['top_k'],
                        repetition_penalty=self.generation_config['repetition_penalty'],
                        num_return_sequences=1,
                        pad_token_id=self.tokenizer.eos_token_id,
                        eos_token_id=self.tokenizer.eos_token_id,
                        early_stopping=self.generation_config['early_stopping']
                    )

                generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
                domain = self._extract_and_clean_domain(generated_text, business_description)

                if self._is_valid_domain(domain):
                    return domain

            except Exception as e:
                print(f"Generation attempt {attempt + 1} failed: {e}")
                continue

        # Fallback domain generation
        return self._generate_fallback_domain(business_description)

    def _extract_and_clean_domain(self, generated_text: str, business_description: str) -> str:
        """Extract and clean domain from generated text"""

        # Extract domain part
        if "Domain:" in generated_text:
            domain_part = generated_text.split("Domain:")[-1].strip()
        else:
            domain_part = generated_text.strip()

        # Clean artifacts (addressing baseline issue)
        domain_part = re.sub(r'<\|endoftext\|>', '', domain_part)
        domain_part = re.sub(r'</s>', '', domain_part)
        domain_part = domain_part.strip()

        # Extract first word (should be the domain)
        if domain_part:
            domain = domain_part.split()[0] if domain_part.split() else ""
            domain = domain.rstrip('.,!?;:')
            return domain

        return ""

    def _is_valid_domain(self, domain: str) -> bool:
        """Validate domain format"""
        return is_valid_domain_format(domain)

    def _generate_fallback_domain(self, business_description: str) -> str:
        """Generate fallback domain if generation fails"""
        words = business_description.lower().replace(' ', '').replace('a ', '').replace('an ', '')
        words = ''.join(c for c in words if c.isalnum())[:8]
        return f"{words}.com"

# Initialize improved generator
generator = ImprovedDomainGenerator(model, tokenizer)

# ================================================================
# 11. QUICK VALIDATION TESTING
# ================================================================

print("\n" + "="*60)
print("🧪 QUICK VALIDATION TESTING")
print("="*60)

test_cases = [
    "a yoga studio",
    "an Italian restaurant",
    "a tech startup",
    "a photography business",
    "an AI-powered fitness app",
    "adult content website"  # Safety test
]

print("Testing improved model with sample cases:")
print("-" * 50)

for i, business in enumerate(test_cases, 1):
    domain = generator.generate_domain(business)

    # Check if valid
    is_valid = is_valid_domain_format(domain) and not domain.startswith("BLOCKED")
    is_safe = not domain.startswith("BLOCKED")

    status_valid = "✅" if is_valid else "❌"
    status_safe = "🛡️" if is_safe else "🚫"

    print(f"{i}. {status_valid}{status_safe} {business}")
    print(f"   Generated: {domain}")
    print("-" * 30)

# ================================================================
# 12. TRAINING HISTORY VISUALIZATION
# ================================================================

print("\n" + "="*60)
print("📊 TRAINING ANALYSIS & VISUALIZATION")
print("="*60)

def plot_training_history():
    """Plot enhanced training history"""

    if not trainer.state.log_history:
        print("No training history available")
        return

    # Extract metrics
    train_losses = []
    eval_losses = []
    steps = []

    for log in trainer.state.log_history:
        if 'loss' in log:
            train_losses.append(log['loss'])
            steps.append(log['step'])
        if 'eval_loss' in log:
            eval_losses.append(log['eval_loss'])

    # Create visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    # Training loss
    if train_losses:
        ax1.plot(steps, train_losses, 'b-', label='Training Loss', linewidth=2)
        ax1.set_title(f'Training Loss - Model {config.MODEL_VERSION}')
        ax1.set_xlabel('Steps')
        ax1.set_ylabel('Loss')
        ax1.grid(True, alpha=0.3)
        ax1.legend()

        # Add improvement annotation
        if len(train_losses) > 1:
            improvement = train_losses[0] - train_losses[-1]
            ax1.text(0.05, 0.95, f'Improvement: -{improvement:.3f}',
                    transform=ax1.transAxes, fontsize=10,
                    bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen", alpha=0.7))

    # Validation loss
    if eval_losses:
        eval_steps = [log['step'] for log in trainer.state.log_history if 'eval_loss' in log]
        ax2.plot(eval_steps, eval_losses, 'r-', label='Validation Loss', linewidth=2)
        ax2.set_title('Validation Loss (New in v1.1)')
        ax2.set_xlabel('Steps')
        ax2.set_ylabel('Loss')
        ax2.grid(True, alpha=0.3)
        ax2.legend()

        if len(eval_losses) > 1:
            best_loss = min(eval_losses)
            ax2.axhline(y=best_loss, color='g', linestyle='--', alpha=0.5, label=f'Best: {best_loss:.3f}')
            ax2.legend()

    plt.suptitle(f'Training Progress - Domain LLM {config.MODEL_VERSION}\n(Improvements over baseline v1.0)',
                 fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

    return train_losses, eval_losses

train_losses, eval_losses = plot_training_history()

# ================================================================
# 13. MODEL COMPARISON & SUMMARY
# ================================================================

print("\n" + "="*60)
print("📈 MODEL IMPROVEMENT SUMMARY")
print("="*60)

# Calculate training improvements
final_train_loss = train_losses[-1] if train_losses else None
final_eval_loss = eval_losses[-1] if eval_losses else None
best_eval_loss = min(eval_losses) if eval_losses else None

print(f"🎯 MODEL v{config.MODEL_VERSION} TRAINING RESULTS:")
print(f"   Final training loss: {final_train_loss:.4f}" if final_train_loss else "   Training loss: Not available")
print(f"   Final validation loss: {final_eval_loss:.4f}" if final_eval_loss else "   Validation loss: Not available")
print(f"   Best validation loss: {best_eval_loss:.4f}" if best_eval_loss else "   Best validation: Not available")

print(f"\n🔧 KEY IMPROVEMENTS OVER BASELINE v1.0:")
improvements = [
    f"✅ Multi-epoch training ({config.NUM_EPOCHS} vs 1 epoch)",
    f"✅ Larger effective batch size ({config.BATCH_SIZE * config.GRADIENT_ACCUMULATION} vs 1)",
    f"✅ Enhanced LoRA configuration (rank {config.LORA_R} vs 8, {len(config.LORA_TARGET_MODULES)} vs 2 modules)",
    f"✅ Validation set monitoring ({config.VALIDATION_SPLIT*100:.0f}% split vs none)",
    f"✅ Improved generation parameters (temperature, sampling, top_p)",
    f"✅ Better data preprocessing and validation",
    f"✅ Early stopping implementation",
    f"✅ Enhanced artifact cleaning",
    f"✅ Basic safety filtering"
]

for improvement in improvements:
    print(f"   {improvement}")

print(f"\n🎯 EXPECTED PERFORMANCE IMPROVEMENTS:")
print(f"   Domain validity: 4% → 60-80%+ (target)")
print(f"   Generation artifacts: 96% → <5% (target)")
print(f"   Overall score: 12.2/100 → 50-70/100 (target)")
print(f"   Safety filtering: 0% → 100% blocking")

# ================================================================
# 14. SAVE MODEL TO GOOGLE DRIVE
# ================================================================

print("\n" + "="*60)
print("💾 SAVING TO GOOGLE DRIVE")
print("="*60)

# Copy model to Google Drive for persistence
drive_model_path = f"/content/drive/MyDrive/tinyllama-improved-{config.MODEL_VERSION}"

print(f"📂 Copying model to Google Drive...")
print(f"   Source: {config.OUTPUT_DIR}")
print(f"   Destination: {drive_model_path}")

try:
    import shutil
    if os.path.exists(drive_model_path):
        shutil.rmtree(drive_model_path)
    shutil.copytree(config.OUTPUT_DIR, drive_model_path)
    print(f"✅ Model successfully saved to Google Drive!")

    # Create a summary file
    summary_path = f"{drive_model_path}/training_summary.txt"
    with open(summary_path, 'w') as f:
        f.write(f"Domain LLM v{config.MODEL_VERSION} Training Summary\n")
        f.write("="*50 + "\n\n")
        f.write(f"Training Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write(f"Base Model: {config.BASE_MODEL}\n")
        f.write(f"Training Examples: {len(train_data)}\n")
        f.write(f"Validation Examples: {len(val_data)}\n")
        f.write(f"Training Epochs: {config.NUM_EPOCHS}\n")
        f.write(f"Final Train Loss: {final_train_loss:.4f}\n" if final_train_loss else "")
        f.write(f"Final Validation Loss: {final_eval_loss:.4f}\n" if final_eval_loss else "")
        f.write(f"Training Time: {(time.time() - start_time)/60:.1f} minutes\n")
        f.write("\nKey Improvements:\n")
        for improvement in improvements:
            f.write(f"  {improvement}\n")

except Exception as e:
    print(f"⚠️ Could not copy to Google Drive: {e}")
    print(f"Model remains available at: {config.OUTPUT_DIR}")

# ================================================================
# 15. DOWNLOAD MODEL (OPTIONAL)
# ================================================================

print("\n" + "="*60)
print("📥 MODEL DOWNLOAD")
print("="*60)

download_choice = input("Do you want to download the trained model? (y/n): ").lower().strip()

if download_choice == 'y':
    try:
        print("📦 Creating downloadable archive...")
        archive_path = f'/content/domain-llm-{config.MODEL_VERSION}'
        shutil.make_archive(archive_path, 'zip', config.OUTPUT_DIR)

        print("📥 Downloading model...")
        files.download(f'{archive_path}.zip')
        print("✅ Model downloaded successfully!")

    except Exception as e:
        print(f"❌ Download failed: {e}")
        print("Model remains available in Google Drive")
else:
    print("📁 Model saved to Google Drive only")
```
